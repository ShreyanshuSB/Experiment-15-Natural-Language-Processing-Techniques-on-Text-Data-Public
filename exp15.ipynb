```python
# Tokenization
import nltk # Natural Langaueg Tool Kit
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "Inter-Continental Nuclear Air Defence System"
tokens = word_tokenize(text)
print(tokens)
print(type(tokens))
```

```python
# Sentences Tokenization
from nltk.tokenize import sent_tokenize
text = "Python is easy language. It is used widely in data science."
sentences = sent_tokenize(text)
print(sentences)
print(type(sentences))
```

```python
# Stop Word Removal
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
# Steming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing","studies","Coding",]
for w in words:
  print(stemmer.stem(w))
```

```python
# Lemmmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("Graph"))
```

```python
# Part-Of-Speech
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words = word_tokenize("Agni -V Is Inter-Continental Nuclear Air Defence Missile")
pos_tags = pos_tag(words)
print(pos_tags)
print(type(pos_tags))
```

```python
#  Word Frequency Counter
from nltk.probability import FreqDist
words = word_tokenize(text)
fd = FreqDist(words)
print("\n",fd.most_common(),"\n")
```
